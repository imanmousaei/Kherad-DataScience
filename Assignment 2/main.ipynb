```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import time
import datetime
import csv
import sys
```

```python
# load datasets
dataset = pd.read_csv('datasets/owid-covid-data.csv', parse_dates=['date'])
dataset.head(10)
```

```python
dataset.tail(10)
```

```python
dataset.sort_values(by='location')
dataset.sort_values(by='date')
dataset.head(10)
```

```python
dataset.info()
```

```python
# fill null fields with previous valid value
dataset.fillna(method='ffill', inplace=True)

# fill remaining nulls with next valid value
dataset.fillna(method='bfill', inplace=True)
```

```python
dataset.isna().sum()
dataset.head(10)
```

```python
# get rid of redundant columns
dataset.drop(columns=['new_cases_smoothed', 'new_deaths_smoothed'], inplace=True)
dataset.head(10)
```

```python
# dataset.plot(kind='bar', title="1-1")
```

```python
n = 10000
dataset.shape
```

```python
plt.plot(dataset['date'][:n], dataset['new_cases'][:n])
plt.title('Plot 1-1', fontsize = 20)
plt.grid()
plt.legend()
plt.show()
```

```python
# the previous plot is not properly distributed. So we shuffle and take the first n(because my laptop cant handle to plot all rows)
from sklearn.utils import shuffle
dataset = shuffle(dataset)
mini_dataset = dataset.sample(frac=0.1)
```

```python
plt.scatter(mini_dataset['date'], mini_dataset['new_cases'], marker='.')
plt.title('Plot 1-2', fontsize = 20)
plt.grid()
plt.legend()
plt.show()
```

```python
plt.scatter(mini_dataset['date'], mini_dataset['new_vaccinations'], marker='.')
plt.title('Plot 3-1', fontsize = 20)
plt.grid()
plt.legend()
plt.show()
```

```python
plt.scatter(dataset['date'], dataset['total_vaccinations'], marker='.', color='b')
plt.scatter(dataset['date'], dataset['new_cases'], marker='.', color='r')
plt.title('Plot 3-1', fontsize = 20)
plt.grid()
plt.legend()
plt.show()
```
