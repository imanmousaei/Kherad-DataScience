```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import csv
```

```python
# load dataset
dataset_path = 'datasets/AB_NYC_2019.csv'
bnb = pd.read_csv(dataset_path)
bnb.sort_values(by='host_id')
bnb.head()
```

```python
class Record:
    def __init__(row):
    id
    name
    host_id
    host_name
    neighbourhood_group
    neighbourhood
    latitude
    longitude
    room_type
    price
    minimum_nights
    number_of_reviews
    last_review
    reviews_per_month
    calculated_host_listings_count
    availability_365
```

```python
# transform format for plt.plot()
records = x = y1 = y2 = y3 = y4 = []
with open(dataset_path,'r') as csvfile:
    lines = csv.reader(csvfile, delimiter=',')
    
    i = -1
    for row in lines:
        i += 1
        if i==0:
            continue
        if i>10:
            break
        
        x.append(row[2]) # host_id
        y1.append(int(row[9])) # price
        y2.append(int(row[10])) # minimum_nights
        y3.append(int(row[11])) # number_of_reviews
        y4.append(int(row[15])) # availability_365
```

```python
# plt.rcParams["figure.figsize"] = [20, 10]
# plt.rcParams["figure.autolayout"] = True

x = bnb['host_id'][:100]
y = bnb['price'][:100]


plt.plot(x, y, label = "price")
plt.legend()
plt.show()
```

```python
plt.plot(x, y1, label = "price")
plt.plot(x, y2, label = "minimum_nights")
plt.plot(x, y3, label = "number_of_reviews")
plt.plot(x, y4, label = "availability_365")
plt.legend()
plt.show()
```

```python
# plt.plot(x, y3, label = "number_of_reviews")
# plt.plot(x, y4, label = "availability_365")

plt.plot(x, y4, color = 'g', linestyle = 'dashed',
         marker = 'o',label = "availability_365")
  
plt.xticks(rotation = 25)
plt.xlabel('Host id')
plt.ylabel('availability_365')
plt.title('Availabililty Report', fontsize = 20)
plt.grid()
plt.legend()
plt.show()
```
